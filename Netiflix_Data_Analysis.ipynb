```python
# Filter the DataFrame to include only movies
movies_df = df[df['Category'] == 'Movie'].copy()

# Count the number of movies released each year
movie_release_counts = movies_df['release_year'].value_counts().sort_index()

# Visualize the relationship between release year and movie count
plt.figure(figsize=(15, 6))
movie_release_counts.plot(kind='bar')
plt.xlabel("Release Year")
plt.ylabel("Number of Movies")
plt.title("Number of Movies Released on Netflix by Year")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame to include only movies
movies_df = df[df['Category'] == 'Movie'].copy()

# Count the number of movies by country and get the top 10
movie_country_counts = movies_df['Country'].value_counts().head(10)

# Display the result
print("Top 10 Countries with the Highest Number of Movies:")
display(movie_country_counts)
```

```python
# Filter the DataFrame to include only TV shows
tv_shows_df = df[df['Category'] == 'TV Show'].copy()

# Count the number of TV shows by country and get the top 5
tv_show_country_counts = tv_shows_df['Country'].value_counts().head(5)

# Display the result
print("Top 5 Countries with the Highest Number of TV Shows:")
display(tv_show_country_counts)
```

## Analyze content duration by country

### Subtask:
Examine the distribution of content duration (movie minutes and TV show seasons) for the top contributing countries.

```python
# Filter for top 10 countries (using the 'top_10_countries' variable from previous analysis)
top_countries_movies = movie_country_counts.index.tolist()
top_countries_tv_shows = tv_show_country_counts.index.tolist()
top_countries = list(set(top_countries_movies + top_countries_tv_shows))
df_top_countries = df[df['Country'].isin(top_countries)].copy()

# Separate movies and TV shows
movies_top_countries = df_top_countries[df_top_countries['Category'] == 'Movie'].copy()
tv_shows_top_countries = df_top_countries[df_top_countries['Category'] == 'TV Show'].copy()

# Convert duration to numeric for movies
movies_top_countries['Duration'] = movies_top_countries['Duration'].str.replace(' min', '', regex=False).astype(int)

# Convert seasons to numeric for TV shows
tv_shows_top_countries['Duration'] = tv_shows_top_countries['Duration'].str.replace(' Seasons?', '', regex=True).astype(int)

# Analyze movie duration by country
print("Movie Duration Analysis for Top Countries:")
display(movies_top_countries.groupby('Country')['Duration'].describe())

# Analyze TV show seasons by country
print("\nTV Show Seasons Analysis for Top Countries:")
display(tv_shows_top_countries.groupby('Country')['Duration'].describe())

# Visualize movie duration distribution for top countries (optional: select a few key countries)
plt.figure(figsize=(15, 8))
sns.boxplot(x='Country', y='Duration', data=movies_top_countries[movies_top_countries['Country'].isin(['United States', 'India', 'United Kingdom'])]) # Example for top 3
plt.title('Movie Duration Distribution for Selected Top Countries')
plt.xlabel('Country')
plt.ylabel('Duration (minutes)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize TV show seasons distribution for top countries (optional: select a few key countries)
plt.figure(figsize=(15, 8))
sns.countplot(x='Duration', hue='Country', data=tv_shows_top_countries[tv_shows_top_countries['Country'].isin(['United States', 'India', 'United Kingdom'])], palette='viridis') # Example for top 3
plt.title('TV Show Seasons Distribution for Selected Top Countries')
plt.xlabel('Number of Seasons')
plt.ylabel('Number of TV Shows')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

## Explore content duration

### Subtask:
Examine the distribution of movie durations and TV show seasons/episodes.


**Reasoning**:
Filter the DataFrame into movies and TV shows, extract and convert duration/seasons, and then visualize their distributions.



```python
movies_df = df[df['Category'] == 'Movie'].copy()
tv_shows_df = df[df['Category'] == 'TV Show'].copy()

movies_df['Duration'] = movies_df['Duration'].str.replace(' min', '').astype(int)
tv_shows_df['Duration'] = tv_shows_df['Duration'].str.replace(' Seasons', '').str.replace(' Season', '').astype(int)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(movies_df['Duration'], bins=30, kde=True)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')

plt.subplot(1, 2, 2)
sns.countplot(x='Duration', data=tv_shows_df, color=sns.color_palette()[0])
plt.title('Distribution of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Number of TV Shows')

plt.tight_layout()
plt.show()
```

```python
import networkx as nx

# Create a strategy graph
G = nx.Graph()

# Main node
G.add_node("Netflix", size=2000)

# Focus areas
areas = ["Movies", "TV Shows", "Top Genres", "Emerging Countries", "Opportunities"]
for area in areas:
    G.add_node(area, size=1500)
    G.add_edge("Netflix", area)

# Draw the graph
plt.figure(figsize=(10,8))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True,
        node_color="skyblue",
        node_size=[G.nodes[n].get("size",1000) for n in G.nodes],
        font_size=10, font_weight="bold")
plt.title("Netflix Content Strategy Map")
plt.show()
```
